```python
#SEQUENTIAL SEARCH PRAKTIKUM

def sequentialsearch(data, key):
    pos = []
    for i in range(len(data)):
        if data[i]==key:
            pos.append(i+1)
    if len(pos)>0:
        print('data', key, 'sebanyak', len(pos), 'ditemukan di posisi', pos)
    else:
        print('data tidak ditemukan')
    return pos

data = [1, 3, 4, 5, 7, 8, 0, 1, 6, 2]
key = 7
print('Sequential Search')
sequentialsearch(data,key)
```

```python
#BINARY SEARCH PRAKTIKUM
def binarysearch(data, key):
    awal = 1
    akhir = len(data) - 1
    ketemu = False
    
    while (awal<=akhir) and not ketemu:
        tengah = int((awal+akhir)/2)
        if key == data[tengah]:
            ketemu = True
            print('data', key, 'ditemukan di posisi', tengah)
        elif (key<data[tengah]):
            akhir=tengah-1
        else:
            awal=tengah+1
    if ketemu == False:
        print('data tidak ditemukan')


data = [1, 2, 5, 8, 10, 13, 15, 17, 19, 22, 25, 29, 30, 35, 40]
key = 29
print('Binary Search')
binarysearch(data, key)
```

```python
#1. MODIFIKASI MENAMPILKAN JUMLAH ITERASI UNTUK MENCARI DATA
#SEQUENTIAL SEARCH : 

def sq_search(data, key):
    pos = []
    iterasi = 0
    for i in range(len(data)): 
        iterasi += 1
        if data[i] == key:
            pos.append(i)

    if len(pos) > 0:
        print('Data', key, 'sebanyak', len(pos), 'ditemukan di posisi', pos)
        print('Jumlah iterasi :', iterasi, '(sequential search)')
    else:
        print('Data', key, 'tidak ditemukan')
        print('Jumlah iterasi :', iterasi, '(sequential search)')
    return pos
    
data = [1, 3, 4, 5, 7, 8, 0, 1, 6, 2]
key = 7
print('=== SEQUENTIAL SEARCH ===')
sq_search(data,key)


#BINARY SEARCH :
def bin_search(data, key):
    awal = 0
    akhir = len(data) - 1
    ketemu = False
    iterasi = 0
    while awal <= akhir and not ketemu :
        iterasi += 1
        tengah = (awal + akhir) // 2
        if data[tengah] == key:
            ketemu = True
            print('Data', key, 'ditemukan di posisi', tengah)
            print('Jumlah iterasi :', iterasi, '(binary search)')
        elif key < data[tengah]:
            akhir = tengah - 1
        else :
            awal = tengah + 1
    if not ketemu: 
        print('Data', key, 'tidak ditemukan')
        print('Jumlah iterasi :', iterasi, '(binary search)')

data = [1, 2, 5, 8, 10, 13, 15, 17, 19, 22, 25, 29, 30, 35, 40]
key = 29
print("=== BINARY SEARCH ===")
bin_search(data, key)
```

```python
#3. CARI HURUF I DALAM KATA 'SISTEM INFORMASI'
Kata = list('sistem informasi')

#SEQUENTIAL SEARCH : 
print('=== SEQUENTIAL SEARCH ===')
sq_search(Kata, 'i')

#BINARY SEARCH : 
print('=== BINARY SEARCH ===')
sorted_kata = sorted(Kata)
bin_search(sorted_kata, 'i')
```
