```python
#PRESENSI KEHADIRAN

siswa = ["Erika", "Dini", "Dio", "Haidar", "Abdul"]  
hadir = 0  
absen = 0  

for nama in siswa:
    print(nama)
    jawaban = input("Apakah hadir? (Y/N): ").upper()

    if jawaban == "Y":  
        hadir += 1  
    elif jawaban == "N":  
        absen += 1  

print(f"Total Siswa = {len(siswa)}")
print(f"Jumlah hadir = {hadir}")  
print(f"Jumlah tidak hadir = {absen}")
```

```python
#KALKULATOR INVESTASI
#KALKULATOR PENGEMBALIAN INVESTASI 

def calculate_profit_and_roi():
    print("Kalkulator Pengembalian Investasi")
    print("---------------------------------")
    
#INPUT DANA YANG DIINVESTASIKAN : 
    while True:
        try:
            total_costs_input = input("Masukkan jumlah total yang diinvestasikan/dibelanjakan (Total Biaya): ")
            total_costs = float(total_costs_input)
            if total_costs < 0:
                print("Total biaya tidak boleh negatif. Silakan masukkan angka positif.")
            else:
                break
        except ValueError:
            print("Input tidak valid. Harap masukkan angka untuk total biaya.")

#INPUT TOTAL PEROLEHAN DARI INVESTASI :
    while True:
        try:
            total_sales_input = input("Masukkan jumlah total yang diperoleh (Total Penjualan): ")
            total_sales = float(total_sales_input)
            if total_sales < 0:
                print("Total penjualan tidak boleh negatif. Silakan masukkan angka positif.")
            else:
                break
        except ValueError:
            print("Input tidak valid. Harap masukkan angka untuk total penjualan.")

#MENGHITUNG UNTUNG/RUGI :
    profit = total_sales - total_costs

    print("\n--- Hasil Perhitungan ---")
    print(f"Total Biaya: Rp {total_costs:,.2f}")
    print(f"Total Penjualan: Rp {total_sales:,.2f}")

 #MENAMPILKAN HASIL UNTUNG/RUGI : 
    if profit > 0:
        print(f"Laba: Rp {profit:,.2f} (Untung!)")
    elif profit < 0:
        print(f"Rugi: Rp {abs(profit):,.2f} (Rugi!)")
    else:
        print("Laba/Rugi: Rp 0.00 (Titik impas!)")


#KALKULATOR ROI : 
    if total_costs == 0:
        print("Tidak dapat menghitung ROI karena total biaya adalah nol.")
    else:
        roi = (profit / total_costs) * 100
        print(f"Pengembalian Investasi (ROI): {roi:.2f}%")

#MENJALANKAN KALKULATOR : 
if __name__ == "__main__":
    calculate_profit_and_roi()
```

```python
def catat_presensi():
    """Mencatat kehadiran siswa di awal pelajaran."""

    daftar_siswa = []
    print("Masukkan nama siswa satu per satu. Ketik 'selesai' jika sudah selesai.")
    while True:
        nama_siswa = input("Nama siswa: ").strip()
        if nama_siswa.lower() == 'selesai':
            break
        if nama_siswa: # Pastikan input tidak kosong
            daftar_siswa.append(nama_siswa)
        else:
            print("Nama siswa tidak boleh kosong.")

    if not daftar_siswa:
        print("Tidak ada siswa dalam daftar kelas. Program berakhir.")
        return

    hadir = 0
    tidak_hadir = 0
    
    print("\n--- Pencatatan Kehadiran ---")
    for siswa in daftar_siswa:
        while True:
            kehadiran = input(f"Apakah {siswa} hadir? (y/n): ").lower().strip()
            if kehadiran == 'y':
                hadir += 1
                break
            elif kehadiran == 'n':
                tidak_hadir += 1
                break
            else:
                print("Input tidak valid. Harap masukkan 'y' untuk hadir atau 'n' untuk tidak hadir.")

    total_siswa = len(daftar_siswa)

    print("\n--- Ringkasan Kehadiran ---")
    print(f"Total siswa di kelas: {total_siswa}")
    print(f"Jumlah siswa hadir: {hadir}")
    print(f"Jumlah siswa tidak hadir: {tidak_hadir}")

# Panggil fungsi untuk menjalankan program
catat_presensi()
```
